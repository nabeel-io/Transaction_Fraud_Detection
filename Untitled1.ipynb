```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_pickle("data/df_reduced.pickle")
```

```python
df
```

```python
import xgboost as xgb
params = {
    'eta':0.1,
    'gamma':0.1,
    'max_depth':10,
    'subsample':0.5,
    'lambda':0.1,
    'alpha':0.5}
```

```python
xgb_model = xgb.XGBClassifier(objective="binary:logistic", eval_metric="error")
xgb_model.fit(df.iloc[:,:-1], df["target"])
```

```python
test = pd.read_pickle("data/test_reduced.pickle")
ypred = xgb_model.predict(test)
```

```python
ypred
```

```python
ids = pd.read_csv("data/test.csv")
```

```python
ID_code = ids["ID_code"]
submission = pd.DataFrame({"ID_code":ID_code, "target":ypred})
submission.to_csv("submission.csv", index=False)
```
